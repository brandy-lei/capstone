```python
# create new feature newtime

from datetime import datetime

fin = open('sensorswww_data.txt','r')

newtime = []
n =0 
for l in fin:
    l=json.loads(l)
    try:
        
        ct=datetime.fromtimestamp(l['time']/1000.0)
        ct = datetime.date(ct)  #  only keep the date
        newtime.append(ct)
    except:
        print l
        break
    # print data
   

fin.close()   

newtime = pd.DataFrame(newtime)
result['newtime'] = newtime
```

## Check the missing ratio for each feature

```python
missing_df = result.isnull().sum(axis=0).reset_index()
missing_df.columns = ['column_name','missing_count']
missing_df['missing_ratio'] = missing_df['missing_count'] / result.shape[0]
```

```python
# select the feature has missing less than 0.9
missing_df.ix[missing_df['missing_ratio']<=0.9]
```
